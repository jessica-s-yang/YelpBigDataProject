# Yelp - Regression  

```python
import numpy as np
import pandas as pd
import csv
import matplotlib.pyplot as plt
from __future__ import division
%matplotlib inline
```

### Setup

Setup code to load the data. This data contains stars ratings and usefulness rating for restaurants. 

```python
reviews = pd.read_csv("review_sample.csv")
```

#### Star vs Useful Linear Regression
First plot a scatter plot for Stars against Useful, and then plot a red linear regression line through the data.

```python
reviews.plot.scatter(x='stars', y='useful')
a,b = np.polyfit(reviews.stars, reviews.useful, 1) # Regression line is y = ax + b
x1 = min(reviews.stars)
x2 = max(reviews.stars)
plt.plot([x1,x2], [a*x1 + b, a*x2 + b], color='red')
plt.show()
```

```python
reviews.plot.scatter(x='useful', y='stars')
a,b = np.polyfit(reviews.useful, reviews.stars, 1) # Regression line is y = ax + b
x1 = min(reviews.useful)
x2 = max(reviews.useful)
plt.plot([x1,x2], [a*x1 + b, a*x2 + b], color='red')
plt.show()
```

```python
# Correlation coefficients (r values)
cc = np.corrcoef(reviews.useful, reviews.stars)[1,0]
print 'Correlation coefficient for stars vs useful :', cc
```
